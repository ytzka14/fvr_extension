```python
import pandas as pd

bnc_coca_list = pd.read_excel("BNC_COCA_lists.xlsx")
```

```python
print(bnc_coca_list)
```

```python
word_list = {}
for i in range(1, 26):
  word_list[str(i)+"k"] = []

for index, row in bnc_coca_list.iterrows():
  words = row['Related forms'].split(", ")
  for word in words:
    raw_word = word.split(" ")[0]
    word_list[row['List ']].append(raw_word)
```

```python
for i in range(1, 26):
  print(str(i)+"k: " + str(word_list[str(i)+"k"]))
```

```python
import json

with open('word_list.json', 'w', encoding='utf-8') as f:
  json.dump(word_list, f, ensure_ascii=False, indent=2)
```

```python
with open('output_with_sentence_counts.json', 'r', encoding='utf-8') as f:
    output_with_sentence_counts = json.load(f)
robot_readers_results = []

for article in output_with_sentence_counts:
  words = list(map(lambda x: "".join([c if c.isalpha() else "" for c in x]), article['GPT_Result'].split(" ")))
  cnts = [0 for _ in range(25)]
  for word in words:
    for i in range(1, 26):
      if word in word_list[str(i)+"k"]:
        cnts[i-1] += 1
  robot_readers_results.append({
    "ArticleID": article['ArticleID'],
    "GPT_Result": article['GPT_Result'],
    "Words": len(words),
    "1k_count": sum(cnts[1:]),
    "2k_count": sum(cnts[2:]),
    "3k_count": sum(cnts[3:]),
    "4k_count": sum(cnts[4:]),
    "5k_count": sum(cnts[5:]),
    "6k_count": sum(cnts[6:]),
    "7k_count": sum(cnts[7:]),
    "8k_count": sum(cnts[8:]),
    "9k_count": sum(cnts[9:]),
    "1k_percent": sum(cnts[1:]) / len(words) * 100 if len(words) > 0 else 0,
    "2k_percent": sum(cnts[2:]) / len(words) * 100 if len(words) > 0 else 0,
    "3k_percent": sum(cnts[3:]) / len(words) * 100 if len(words) > 0 else 0,
    "4k_percent": sum(cnts[4:]) / len(words) * 100 if len(words) > 0 else 0,
    "5k_percent": sum(cnts[5:]) / len(words) * 100 if len(words) > 0 else 0,
    "6k_percent": sum(cnts[6:]) / len(words) * 100 if len(words) > 0 else 0,
    "7k_percent": sum(cnts[7:]) / len(words) * 100 if len(words) > 0 else 0,
    "8k_percent": sum(cnts[8:]) / len(words) * 100 if len(words) > 0 else 0,
    "9k_percent": sum(cnts[9:]) / len(words) * 100 if len(words) > 0 else 0
  })
```

```python
with open('robot_readers_results.json', 'w', encoding='utf-8') as f:
    json.dump(robot_readers_results, f, ensure_ascii=False, indent=2)
```
